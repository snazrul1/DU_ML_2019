```python
#Simple class and polymorphisms



#Base parent class
class Person(object):
    """This is a person class"""
    
    def __init__(self, name, age):
        """Constructor with self handle"""
        self.name = name
        self.age = age
        
    def get_name(self):
        return self.name
        
    def get_age(self):
        return self.age

    

#Inheriting child class
class Student(Person):
    """Student inherits from Person"""
    
    def __init__(self, name, age, school):
        #Get Person's attributes
        Person.__init__(self, name, age)
        #Unique school info
        self.school = school
        
    def get_school(self):
        return self.school
        


#Initialize class as object
hammad = Student("Hammad", 7, "SD School")

#Person attributes
print(hammad.get_name())
print(hammad.get_age())

#Student attributes
print(hammad.get_school())
```

# Numpy

Numpy is a math library that runs on C for faster implementation. It's very useful for solving large matrix operations.

```python
import numpy as np
```

```python
print(np.sin(0.243))
print()

print(np.pi)
print()
```

```python
mat = np.matrix([[1,2,3],[4,5,6],[7,8,9]])
print(mat)
print()

print(np.cross(mat, mat))
print()

print(np.dot(mat,mat))
print()

print(np.linalg.inv(mat))
```

# Matplotlib
Matplotlib is the Python plotting library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
# Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# Creates two subplots and unpacks the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Creates four polar axes, and accesses them through the returned array
fig, axes = plt.subplots(2, 2, subplot_kw=dict(polar=True))
axes[0, 0].plot(x, y)
axes[1, 1].scatter(x, y)
```

# Pandas
Pandas is a data management library and will be one of the most important library for a data scientist

```python
import pandas as pd
```

```python
# The inital set of baby names and birth rates
names = ['Bob','Jessica','Mary','John','Mel']
births = [68, 15, 77, 78, 73]
```

```python
BabyDataSet = list(zip(names,births))
print(BabyDataSet)
```

```python
df = pd.DataFrame(data = BabyDataSet, columns=['Names', 'Births'])
df.head()
```

```python
print(df["Births"].max())
```

```python
df.sort_values("Births", ascending=False)
```

```python
df["Log_Births"] = df["Births"].apply(lambda x: np.log(x))
df.head()
```
